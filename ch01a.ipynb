## 펭균 데이터

```python
penguines = pd.read_csv("data/penguins.csv")
penguines = penguines.dropna() #빈값을 지울께 #전처리를 빠르게
penguines.info()
```

```python
#type(penguines)
penguines.keys() #타입에러경우 해결 []
```

```python
X = penguines[["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]]
y = penguines["species"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42) 
```

```python
from sklearn.neighbors import KNeighborsClassifier #분류모델
knn = KNeighborsClassifier()
knn.fit(X_train, y_train) #학습할꺼야. #타겟은 숫자가 아니여도 결과가 나온다.
```

```python
from sklearn.metrics import accuracy_score
y_pred = knn.predict(X_test) #예측할꺼야.
accuracy = accuracy_score(y_test, y_pred) #예측퍼센트 결과 0.75 
accuracy
```

## 성능을 높여보자 결과:0.9880952380952381 전처리해서 늘었다. (전: 0.75)

```python
#성능을 높여보자 결과:0.9880952380952381 전처리해서 늘었다. 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() #StandardScaler가져와서 객체를 만들었다.
X_train_scaled = scaler.fit_transform(X_train) #fit 은 무조건 train값 넣어야한다. (test값안됨)
X_test_scaled = scaler.transform(X_test)
```

```python
knn.fit(X_train_scaled, y_train)
```

```python
y_pred = knn.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)

accuracy
```

```python
k_values = [1,3,5,7,9,11]
accuracies = []
for k in k_values:
    #1. k에 따른 모델을 학습
    knn_temp = KNeighborsClassifier(n_neighbors=k)
    knn_temp.fit(X_train_scaled, y_train)
    #2. 예측
    y_pred_temp = knn_temp.predict(X_test_scaled) 
    acc = accuracy_score(y_test, y_pred_temp)
    accuracies.append(acc)
    print(f"k={k}: 정확도:{acc:.4f}")
    #3. 정확도 측정
    #4. 저장
    #5. 출력
best_k = k_values[np.argmax(accuracies)] #가장 정확도가 높았던 k 값을 찾는 코드
#최적의 k를 찾으세요
```

```python
best_k #
```

```python
#과적합
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train_scaled, y_train)
y_pred = knn.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
